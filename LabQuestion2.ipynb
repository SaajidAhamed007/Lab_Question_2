```python
import random
import matplotlib.pyplot as plt

# Modified linear search that counts comparisons
def linear_search_with_comparisons(arr, key):
    comparisons = 0
    for i in range(len(arr)):
        comparisons += 1
        if arr[i] == key:
            return i, comparisons
    return -1, comparisons

# Input sizes to test
input_sizes = [100, 500, 1000, 5000, 10000]

# Store results
best_case_comparisons = []
average_case_comparisons = []
worst_case_comparisons = []

for n in input_sizes:
    arr = [random.randint(0, 100000) for _ in range(n)]
    
    # Best case: key at first index
    _, best = linear_search_with_comparisons(arr, arr[0])
    best_case_comparisons.append(best)
    
    # Average case: key in the middle
    _, avg = linear_search_with_comparisons(arr, arr[n // 2])
    average_case_comparisons.append(avg)
    
    # Worst case: key not in list
    _, worst = linear_search_with_comparisons(arr, -1)
    worst_case_comparisons.append(worst)

# Print results in tabular form
print(f"{'Input Size':<12}{'Best':<10}{'Average':<10}{'Worst':<10}")
for i in range(len(input_sizes)):
    print(f"{input_sizes[i]:<12}{best_case_comparisons[i]:<10}{average_case_comparisons[i]:<10}{worst_case_comparisons[i]:<10}")

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(input_sizes, best_case_comparisons, marker='o', label='Best Case (O(1))')
plt.plot(input_sizes, average_case_comparisons, marker='s', label='Average Case (O(n/2))')
plt.plot(input_sizes, worst_case_comparisons, marker='^', label='Worst Case (O(n))')

plt.title('Number of Comparisons in Linear Search')
plt.xlabel('Input Size (n)')
plt.ylabel('Number of Comparisons')
plt.legend()
plt.grid(True)
plt.show()

# Observations
print("\nObservation:")
print(" - Best case always results in 1 comparison (constant time).")
print(" - Average case grows linearly but roughly half the input size.")
print(" - Worst case grows linearly with n, showing O(n) behavior.")
```
